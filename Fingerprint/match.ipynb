```
#매칭률 구하는 함수 작성해보기!
import csv
import math

def match(fname1, fname2, mth):                                                 #파일 2개 불러와 mth에 따라 매칭률 구하는 함수
  mlist1=[]
  mlist2=[]
  f1=open(fname1, "r")
  rdr1=csv.reader(f1)
  for line in rdr1:
    mlist1.append(line)
  f1.close()

  f2=open(fname2, "r")
  rdr2=csv.reader(f2)
  for line in rdr2:
    mlist2.append(line)
  f2.close()                                                                    #파일 2개를 읽어 각각 mlist1, mlist2에 저장

  mcnt=0
  for i in range(0, len(mlist1)):
    for j in range(0, len(mlist2)):
      d=math.sqrt(((int(mlist1[i][0])-int(mlist2[j][0]))**2)+((int(mlist1[i][1])-int(mlist2[j][1]))**2))
      if(d<=mth):
        mcnt+=1                                                                 #각 원소의 거리값을 구해 mth보다 작으면 매칭카운트에 1 더함

  mr=mcnt/(len(mlist1)*len(mlist2))*100                                         #최종적으로 구한 매칭카운트에 2개의 배열 원소 개수로 나눠 매칭률 구하기
  return mr
```

```
#Mth와 Dth, 지문 종류에 따른 FRR 구하는 함수 만들기!
def FRR(mth, dth, num):                                                         #mth와 dth, 지문번호를 입력받아 FRR을 출력해주는 함수!
  FRR=0
  i=num-1                                                                       #입력받은 지문번호 ex)1, 2, ..., 8 에서 1을 빼준다(인덱스 때문에)
  for j in range(10*i, 10*i+2):                                               
    dcnt=0
    for k in range(10*i+2, 10*(i+1)):                                           #(1), (2)와 (3)~(10)을 비교한다
      mr=match(mnt_fname[j], mnt_fname[k], mth)                                 #매칭률 구하는 함수를 써서 해당 파일의 매칭률을 구해주고
      if(mr>=dth):                                                              #매칭률이 dth값보다 크면 통과이므로
        dcnt+=1                                                                 #카운트를 1 증가시킨다.
    FRR+=(8-dcnt)/8*100                                                         #거부율을 찾는 것이므로 8에서 dcnt값을 빼고 8로 나눠 FRR에 누적합해준다
  FRR=FRR/2                                                                     #총 비교 횟수인 2으로 나눠 FRR을 구한다.
  return FRR
```

```
mth=[10, 50, 100]                                                               #모든 mth와 dth 값에 따라 반복문을 이용해 FRR값을 출력한다
dth=[10, 30, 50, 70, 90]
for i in mth:
  for j in dth:
    print(f"mth={i}, dth={j}, FRR 값")
    a=0
    for k in range(1, 9):
      print(f"{FRR(i, j, k)}  ", end='')
      a+=FRR(i, j, k)

    a=a/8
    print(f"\n전체 지문의 FRR 평균 값 : {a}\n")
```

```
#mth, dth, 지문 종류에 따른 FAR 구하는 함수!

def FAR(mth, dth, num):                                                         #mth와 dth, 그리고 지문번호를 입력받아 FAR을 출력해주는 함수
  FAR=0
  i=num-1                                                                       #입력받은 지문번호 ex)1, 2, ..., 8 에서 1을 빼준다(인덱스 때문에)
  for j in range(10*i, 10*i+2):                                                 #(1), (2)번 지문을 자기 것이 아닌 (3)~(10) 총 56개와 비교
    dcnt=0
    for n in range(0, 8):
      if(i==n):
        continue                                                                #다른 종류의 지문과 비교해야하므로 같은 지문인 경우 패쓰하기                            
      for k in range(10*n+2, 10*(n+1)): 
        mr=match(mnt_fname[j], mnt_fname[k], mth)                               #mth에 따른 매칭률을 구하고
        if(mr>=dth):                                                            #매칭률이 dth보다 크면 통과이므로 dcnt에 1을 증가시킨다
          dcnt+=1
    FAR+=dcnt/56*100                                                            #총 비교 횟수가 56번이므로 56으로 나눠 FAR에 누적합해준다.

  FAR=FAR/2                                                                     #(1), (2)이므로 2로 나눠준다
  return FAR
```

```
mth=[10, 50, 100]                                                               #모든 mth와 dth 값에 따라 반복문을 이용해 FAR값을 출력한다
dth=[10, 30, 50, 70, 90]
print("<FAR>")
for i in mth:
  for j in dth:
    print(f"mth={i}, dth={j}")
    a=0
    for k in range(1, 9):
      print(f"{FAR(i, j, k)}  ", end='')
      a+=FAR(i, j, k)

    a=a/8
    print(f"\n전체 지문의 FAR 평균 값 : {a}\n")
```

```
# mth에 따른 identification 매칭비율 구하는 함수!

def identification(mth):                                                        #mth를 인자로 받아 identification의 매칭비율(MR)을 출력해주는 함수
  maxlist=[]
  for i in range(0, 8):                                                         #총 8개의 지문에서
    for k in range(10*i, 10*i+2):                                               # 테스트용인 (1)과 (2)를
      index=0
      maxmat=0
      for n in range(0, 8):
        for j in range(10*n+2, 10*(n+1)):                                       # 1 (3)~1 (10)에서부터 8 (3)~8 (10)까지 64번 비교
          mr=match(mnt_fname[k], mnt_fname[j], mth)                             #mth에 따라 매칭률을 구한 후
          if(maxmat<mr):
            maxmat=mr
            index=j
      info=[k, maxmat, index]                                                   #매칭률이 가장 최대일 때 그 때의 테스트용 지문 ID와 비교용 지문 ID 저장
      maxlist.append(info)                                                      #리스트에 저장

  cnt=0
  for list in maxlist:
    str1=mnt_fname[list[0]].split(' ')                                          #지문 파일 형식이 1 (1).csv이므로 띄어쓰기를 기준으로 문자열을 자르고
    str2=mnt_fname[list[2]].split(' ')
    if(str1[0]==str2[0]):                                                       #문자열에서 첫번째로 오는 숫자가 그 지문의 ID이므로 비교한다.
      cnt+=1                                                                    #ID가 동일한 경우 카운트 증가
  MR=cnt/16*100                                                                 #총 카운트에서 16개의 테스트용 지문을 비교했으므로 이로 나눠 비율 계산
  NMR=(16-cnt)/16*100
  print(f"매칭비율 MR: {MR}%")  
  print(f"매칭안된비율 NMR: {NMR}%")  
```

```
print("Mth=10")
identification(10)
print("\nMth=50")
identification(50)
print("\nMth=100")
identification(100)
```
