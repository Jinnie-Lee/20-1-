```
#함수 이용해 모든 지문 파일 '세선화'까지하고 값 배열에 저장하기!!
import matplotlib.image as mpimg
import matplotlib.pylab as plt
import numpy as np
import matplotlib.pylab as pylab
import skimage.morphology
from skimage.morphology import skeletonize
from skimage import filters, feature, img_as_float
from skimage.io import imread

list_sklt=[]
file_name=['1 (1).bmp', '1 (2).bmp', '1 (3).bmp', '1 (4).bmp', '1 (5).bmp', '1 (6).bmp', '1 (7).bmp', '1 (8).bmp', '1 (9).bmp', '1 (10).bmp',
           '2 (1).bmp', '2 (2).bmp', '2 (3).bmp', '2 (4).bmp', '2 (5).bmp', '2 (6).bmp', '2 (7).bmp', '2 (8).bmp', '2 (9).bmp', '2 (10).bmp', 
           '3 (1).bmp', '3 (2).bmp', '3 (3).bmp', '3 (4).bmp', '3 (5).bmp', '3 (6).bmp', '3 (7).bmp', '3 (8).bmp', '3 (9).bmp', '3 (10).bmp', 
           '4 (1).bmp', '4 (2).bmp', '4 (3).bmp', '4 (4).bmp', '4 (5).bmp', '4 (6).bmp', '4 (7).bmp', '4 (8).bmp', '4 (9).bmp', '4 (10).bmp', 
           '5 (1).bmp', '5 (2).bmp', '5 (3).bmp', '5 (4).bmp', '5 (5).bmp', '5 (6).bmp', '5 (7).bmp', '5 (8).bmp', '5 (9).bmp', '5 (10).bmp', 
           '6 (1).bmp', '6 (2).bmp', '6 (3).bmp', '6 (4).bmp', '6 (5).bmp', '6 (6).bmp', '6 (7).bmp', '6 (8).bmp', '6 (9).bmp', '6 (10).bmp', 
           '7 (1).bmp', '7 (2).bmp', '7 (3).bmp', '7 (4).bmp', '7 (5).bmp', '7 (6).bmp', '7 (7).bmp', '7 (8).bmp', '7 (9).bmp', '7 (10).bmp', 
           '8 (1).bmp', '8 (2).bmp', '8 (3).bmp', '8 (4).bmp', '8 (5).bmp', '8 (6).bmp', '8 (7).bmp', '8 (8).bmp', '8 (9).bmp', '8 (10).bmp', ]

def makeskeleton(fname):
  img=mpimg.imread(fname)                                                       #입력받은 파일이름을 드라이브에서 읽어온다

  img_bin=np.zeros((144,144))
  img_bin[img<127]=1
  img_bin[img>=127]=0                                                           #이진화 진행

  for j in range(0, 144):                                                       
    for i in range(0, 24):
      img_bin[i][j]=0
    for i in range(120, 144):
      img_bin[i][j]=0

  for i in range(24, 120):
    for j in range(0, 24):
      img_bin[i][j]=0
    for j in range(120, 144):
      img_bin[i][j]=0                                                           #외곽부분을 0으로 해준다

  skeleton=skeletonize(img_bin)                                                 #세선화 진행
  list_sklt.append(skeleton)                                                    #세선화 된 값을 리스트에 넣는다

for i in file_name:                                                             #반복문을 이용해 모든 파일 특징점 추출해 리스트에 저장!
  makeskeleton(i)
```

```
#함수 이용해 모든 지문 파일 '열림과 닫힘 후 세선화'까지 하고 값 배열에 저장하기!
from skimage.morphology import binary_opening, binary_closing, disk
import matplotlib.image as mpimg
import matplotlib.pylab as plt
import numpy as np
import matplotlib.pylab as pylab
import skimage.morphology
from skimage.morphology import skeletonize
from skimage import filters, feature, img_as_float
from skimage.io import imread

list_sklt=[]
file_name=['1 (1).bmp', '1 (2).bmp', '1 (3).bmp', '1 (4).bmp', '1 (5).bmp', '1 (6).bmp', '1 (7).bmp', '1 (8).bmp', '1 (9).bmp', '1 (10).bmp',
           '2 (1).bmp', '2 (2).bmp', '2 (3).bmp', '2 (4).bmp', '2 (5).bmp', '2 (6).bmp', '2 (7).bmp', '2 (8).bmp', '2 (9).bmp', '2 (10).bmp', 
           '3 (1).bmp', '3 (2).bmp', '3 (3).bmp', '3 (4).bmp', '3 (5).bmp', '3 (6).bmp', '3 (7).bmp', '3 (8).bmp', '3 (9).bmp', '3 (10).bmp', 
           '4 (1).bmp', '4 (2).bmp', '4 (3).bmp', '4 (4).bmp', '4 (5).bmp', '4 (6).bmp', '4 (7).bmp', '4 (8).bmp', '4 (9).bmp', '4 (10).bmp', 
           '5 (1).bmp', '5 (2).bmp', '5 (3).bmp', '5 (4).bmp', '5 (5).bmp', '5 (6).bmp', '5 (7).bmp', '5 (8).bmp', '5 (9).bmp', '5 (10).bmp', 
           '6 (1).bmp', '6 (2).bmp', '6 (3).bmp', '6 (4).bmp', '6 (5).bmp', '6 (6).bmp', '6 (7).bmp', '6 (8).bmp', '6 (9).bmp', '6 (10).bmp', 
           '7 (1).bmp', '7 (2).bmp', '7 (3).bmp', '7 (4).bmp', '7 (5).bmp', '7 (6).bmp', '7 (7).bmp', '7 (8).bmp', '7 (9).bmp', '7 (10).bmp', 
           '8 (1).bmp', '8 (2).bmp', '8 (3).bmp', '8 (4).bmp', '8 (5).bmp', '8 (6).bmp', '8 (7).bmp', '8 (8).bmp', '8 (9).bmp', '8 (10).bmp', ]

list_opcl=[]

def makeopcl(fname):                                                            #파일 이름을 인자로 받고 이진화, 열림닫힘, 세선화해서 
                                                                                #리스트에 저장해주는 함수
  img=mpimg.imread(fname)                                                       #입력받은 파일이름을 드라이브에서 읽어온다

  img_bin=np.zeros((144,144))
  img_bin[img<127]=1
  img_bin[img>=127]=0                                                           #이진화 진행

  for j in range(0, 144):                                                       
    for i in range(0, 24):
      img_bin[i][j]=0
    for i in range(120, 144):
      img_bin[i][j]=0

  for i in range(24, 120):
    for j in range(0, 24):
      img_bin[i][j]=0
    for j in range(120, 144):
      img_bin[i][j]=0                                                           #외곽부분을 0으로 해준다

  img1=binary_opening(img_bin, disk(1))
  img2=binary_closing(img1, disk(1))                                            #열림과 닫힘 진행

  skeleton=skeletonize(img2)                                                    #세선화 진행
  list_opcl.append(skeleton)                                                    #세선화 된 값을 리스트에 추가



for i in file_name:                                                             #반복문을 이용해 모든 파일 특징점 추출해 리스트에 저장!
  makeopcl(i)
```

```
#특징점 추출하는 함수 만들어 지문 특징점들 파일로 저장~!
import csv
mnt_fname=['1 (1).csv', '1 (2).csv', '1 (3).csv', '1 (4).csv', '1 (5).csv', '1 (6).csv', '1 (7).csv', '1 (8).csv', '1 (9).csv', '1 (10).csv',
           '2 (1).csv', '2 (2).csv', '2 (3).csv', '2 (4).csv', '2 (5).csv', '2 (6).csv', '2 (7).csv', '2 (8).csv', '2 (9).csv', '2 (10).csv', 
           '3 (1).csv', '3 (2).csv', '3 (3).csv', '3 (4).csv', '3 (5).csv', '3 (6).csv', '3 (7).csv', '3 (8).csv', '3 (9).csv', '3 (10).csv', 
           '4 (1).csv', '4 (2).csv', '4 (3).csv', '4 (4).csv', '4 (5).csv', '4 (6).csv', '4 (7).csv', '4 (8).csv', '4 (9).csv', '4 (10).csv', 
           '5 (1).csv', '5 (2).csv', '5 (3).csv', '5 (4).csv', '5 (5).csv', '5 (6).csv', '5 (7).csv', '5 (8).csv', '5 (9).csv', '5 (10).csv', 
           '6 (1).csv', '6 (2).csv', '6 (3).csv', '6 (4).csv', '6 (5).csv', '6 (6).csv', '6 (7).csv', '6 (8).csv', '6 (9).csv', '6 (10).csv', 
           '7 (1).csv', '7 (2).csv', '7 (3).csv', '7 (4).csv', '7 (5).csv', '7 (6).csv', '7 (7).csv', '7 (8).csv', '7 (9).csv', '7 (10).csv', 
           '8 (1).csv', '8 (2).csv', '8 (3).csv', '8 (4).csv', '8 (5).csv', '8 (6).csv', '8 (7).csv', '8 (8).csv', '8 (9).csv', '8 (10).csv', ]
           
def minutiae(skeleton, fname):                                                  #세선화된 값 저장되어 있는 리스트와 파일 이름을 인자로 받아
                                                                                #특징점(단점)을 추출하고 그 위치 인덱스를 인자로 받은 파일 이름을 갖는
                                                                                #파일로 저장하는 함수
  for i in range(20, 120):  
    for j in range(20, 120):                                                    #실제 지문 사진이 존재하는 픽셀이 20~120이므로
      if(skeleton[i][j]==1):                                                    #p값이 1일 때만
        p1=skeleton[i][j+1]*1
        p2=skeleton[i-1][j+1]*1
        p3=skeleton[i-1][j]*1
        p4=skeleton[i-1][j-1]*1
        p5=skeleton[i][j-1]*1
        p6=skeleton[i+1][j-1]*1
        p7=skeleton[i+1][j]*1
        p8=skeleton[i+1][j+1]*1
        Cn=abs(p1-p2)+abs(p2-p3)+abs(p3-p4)+abs(p4-p5)+abs(p5-p6)+abs(p6-p7)+abs(p7-p8)+abs(p8-p1)
        Cn=Cn/2                                                                 #Cn값 계산
        if(Cn==1):                                                              #Cn이 1일 때 = p가 단점일 때
          locate=[i, j]                                                         #p점의 인덱스를 list에 저장
          f=open(fname, 'a', newline='')                                        #파일에 저장한다.
          wr=csv.writer(f)
          wr.writerow(locate)
          f.close()

for i in range(0, 80):                                                          #반복문 돌려서 각 지문별 단점 위치 파일 생성 후 인덱스들 저장
  minutiae(list_opcl[i], mnt_fname[i])
```
