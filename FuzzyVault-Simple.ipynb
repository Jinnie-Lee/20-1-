```
def p_x(x, coeffs):
    # returns p(x) at the given x, where p(x) is a polynomial function defined 
    # by its coefficients.
    y = 0
    #degree = len(coeffs) - 1
    degree = 4
    for coeff in coeffs:        
        y += x**degree * coeff
        degree -= 1
        if degree<0:
            return y
    return y
```

```
degree = 4 # degree 4 polynomial
t = 10     # number of features in each template
r = 40     # number of chaff points
```

```
#encoding
template = template1['Jayme Woogerd']
#print(template)
y = 0
#degree = len(key) - 1
#print(degree)

# a fuzzy vault in which this data is encrypted
vault = []
coeffs = key

# calculate genuine points
for point in template:
    vault.append([point, p_x(point, coeffs)])

#add chaff points
max_x = max(template)
max_y = max([y for [x, y] in vault])

for i in range(t, r):
    x_i = uniform(0, max_x * 1.1)
    #y_i = uniform(0, p_x(max_x, coeffs) * 1.1)
    y_i = uniform(0, max_y * 1.1)
    vault.append([x_i, y_i])
    
shuffle(vault)

print(vault)    
print(len(vault))
print(coeffs)
```

```
def approx_equal(a, b, epsilon):
    #print(a)
    #print(b)
    return abs(a - b) < epsilon
```

```
def project(x):
    for point in vault:
        if (approx_equal(x, point[0], 0.001)):
        #if (approx_equal(x, point[0], 0.01)):
            return [x, point[1]]
    return None
```

```
degree = 4 # degree 4 polynomial

template = template2['Jayme Woogerd']
#template = template2['Ming Chow']
#template = template2['Norman Ramsey']
Q_x = [] 
Q_y = []

#Q = zip(*[ != None])
for point in template:
    cand = project(point)
    if (cand != None):
        Q_x.append(cand[0])
        Q_y.append(cand[1])

print(len(Q_x))
print(Q_x)
print(len(Q_y))
print(Q_y)

coeffs_cand = 0
coeffs_cand = polyfit(Q_x, Q_y, deg=degree)
```
